## **Import the necessary libraries:**

```python
import tensorflow as tf
import numpy as np
from keras.models import Sequential
from tensorflow.keras.layers import Dense
```

## **Prepare Dataset:**

```python
x = np.random.uniform(-10, 10, (10000, 1))
y = 3 * x**3 + 7 * x**2 - 12 * x + 2
```

## **Define the model architecture:**

```python
# Define the model architecture
model = Sequential([
    Dense(100, input_shape=[1], activation='relu'),
    Dense(100, activation='relu'),
    Dense(1)
])

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
history = model.fit(x, y, epochs=500, batch_size=32)
```

## **Answer Prediction:**

```python
# Print result
print('-----------------------RESULT--------------------\n')
numbers = [0, 1, -1, -2, 2, 3, -3]

for x in numbers:
  result = model.predict([x])
  # print("x = ", x ,"  y = ",(result[0][0]),"  original y = ",(3 * x**3 + 7 * x**2 - 12 * x + 2))
  print("\nx = ", x ,"  y = {:.2f}".format(round(result[0][0], 2)),"  original y = ",(3 * x**3 + 7 * x**2 - 12 * x + 2))
  
```
